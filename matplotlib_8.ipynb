```python
import numpy as np
tM = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
tm = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
```

```python
import matplotlib.pyplot as plt

months = np.arange(1,13)
plt.figure(figsize=(8,6))
plt.plot(months, tM, 'ro')
plt.plot(months, tm, 'bo')
plt.xlabel('Months')
plt.ylabel('Min and Max temperature')
```

##Task1: fitting it(above graph) to the periodic function

```python
from scipy import optimize

def yearly_temps(times, avg, ampl, time_offset):
    return (avg
            + ampl * np.cos((times + time_offset) * 2 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(yearly_temps, months, tM, [20, 10, 0])
res_min, cov_min = optimize.curve_fit(yearly_temps, months, tm, [-40, 20, 0])
```

##Task2: plot the fit

```python
days = np.linspace(0, 12, num=365)

plt.figure(figsize=(8,6))
plt.plot(months, tM, 'ro')
plt.plot(days, yearly_temps(days, *res_max), 'r-')
plt.plot(months, tm, 'bo')
plt.plot(days, yearly_temps(days, *res_min), 'b-')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')

plt.show()
```
